```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_csv('data/shopping_data.csv')

basket_df = df.groupby('invoice_no')['category'].apply(list).reset_index()
basket_df
```

```python
# 전처리: '결제 데이터 전처리 실습'의 코드를 여기에 붙여 넣으세요
basket_df = df.groupby('invoice_no')['category'].apply(list).reset_index()
basket_df.head()
```

```python
# TransactionEncoder로 추가 전처리
te = TransactionEncoder()

te_result = te.fit_transform(basket_df['category'])  # 여기에 코드를 작성하세요

te_df = pd.DataFrame(te_result, columns=te.columns_)

te_df.head
```

```python
# Apriori 알고리즘 적용하기
frequent_itemsets = apriori(te_df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
# 연관 규칙 찾기
ar = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.2)
ar
```
