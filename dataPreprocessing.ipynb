```python
# 데이터 불러오기

import pandas as pd
df = pd.read_csv('data/retail_data.csv')
df.head()
```

```python
# 결제 id와 상품 정보만 남기기
df2 = df[['OrderID', 'ProdName']]
df2.head()
```

```python
# groupby() 함수를 이용하여 결제 ID 별로 구매 상품 목록을 묶기
# apply(list)라는 코드를 적용하면, 기준이 되는 컬럼(ProdName)을 리스트 형태로 묶는다는 의미

basket_df = df2.groupby('OrderID')['ProdName'].apply(list).reset_index()
basket_df.head()
```
